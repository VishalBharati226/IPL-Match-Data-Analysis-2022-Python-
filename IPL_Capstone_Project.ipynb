###  Key Player Performances

1 Most "Player of the Match" Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x = count.values, y = count.index,palette='deep')
plt.title("Top 10 players with man of the match")
```

```python
# 2 Top Scorers batsman
high_score = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high_score
```

```python
colors = ['#4CAF50', '#2196F3', '#FFC107', '#E91E63']
high_score.plot(kind='barh', color=colors)
```

```python
# 10 Best Bowling Figures
df['highest_wicket'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wicket'] = df['highest_wicket'].astype(int)
top_bowler = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)
top_bowler.plot(kind = 'barh',color= colors)
```
